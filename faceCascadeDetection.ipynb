```
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import time
from pnslib import utils
```

```
cascPath = '/content/face_detect.xml'
faceCascade = cv2.CascadeClassifier(cascPath)
```

```
# read the image
img = cv2.imread('messi.jpg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```
dir(faceCascade)
```

```
# defining the function to detect the faces
def detect_faces_show(fpath):
  img = cv2.imread(fpath)
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

  # detected faces in the image
  # will return an integer (# of faces)
  faces = faceCascade.detectMultiScale(
      gray, scaleFactor = 1.2,
      minNeighbors=5,
      minSize=(30, 30),
      flags=cv2.CASCADE_SCALE_IMAGE
  )
  print("Found {} faces".format(len(faces)))

  # draw a rectangle around the faces
  for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```
detect_faces_show('messi.jpg')
```

```
img = cv2.imread('/content/Girls Frontline.png')
detect_faces_show('/content/Girls Frontline.png')
```
