```
import pandas as pd

cars = {'brand' : ['Honda', 'toyota', 'ford'],
        'price' : [220000, 110000, 13000]}

cars_df = pd.DataFrame(cars, columns= ['brand', 'price'])
print(cars_df)
```

```
cars_df = pd.DataFrame(cars, columns= ['brand', 'price'], index = ['car1', 'car2', 'car3'])
print(cars_df)
```

```
year = [2010, 2011, 2008]
cars_df['year'] = year
print(cars_df)
```

```
cars_df.insert(1, 'model', ["Civid", "Prius", "Focus"], True) #can duplicate field name
cars_df
```

```
#new_Car_1 = {'brand':'hyundai', 'model':'Avante','year': 2010, 'price':300000}
#cars_df = cars_df.append(new_Car_1, ignore_index = True)

cars_df.loc['car4'] = ['Hyundai', 'Avante', 100000, 2010]
cars_df.loc['car3'] = ['Suzuki', 'Swift', 26000, 2013]
```

```
discount = cars_df['price'] - (0.1 * cars_df['price'])
cars_df['10% discount'] = discount
print(cars_df)
```

```
customer = ["Ali", "Jake", "Alex"]
discount_90 = cars_df['price'] - cars_df['price'] * 0.9
#cars_df.insert(2, "Customer", customer,False)
cars_df.insert(3, "90% discount", discount_90, False)
print(cars_df)
```
